```python
df_interesantes = df_copia.drop(['CASE_NUMBER', 'TYPE', 'AREA', 'LOCATION',
       'ACTIVITY', 'NAME', 'TIME', 'PDF', 'HREF',
       'UNNAMED: 0_x', 'CASE NUMBER', 'SPECIES ', 'INVESTIGATOR OR SOURCE',
       'HREF FORMULA', 'CASE NUMBER.1', 'CASE NUMBER.2', 'ORIGINAL ORDER',
       'UNNAMED:22', 'UNNAMED: 23', 'UNNAMED: 0_y', 'INJURY', 'SEX_UAL'], axis=1)

df_interesantes
```

2. ¿Hay valores duplicados en nuestro dataframe? En caso de que los haya, eliminándlos.

```python
df_interesantes.duplicated().sum()
```

```python
df_interesantes.drop_duplicates(inplace=True)
df_interesantes
```

3. Como hemos visto, algunas columnas no tienen el tipo de datos que deberían. Cambiad el tipo de dato para la columna de year.

```python
df_interesantes["YEAR"] = df_interesantes["YEAR"]. astype(int, errors = "ignore")
df_interesantes
```
